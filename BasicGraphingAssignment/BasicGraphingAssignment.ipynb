```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
import pandas as pd
```

```python
df_OpiodsVA = pd.read_csv('OpidsVA.csv')

"""A scatter plot is the best graph to display this data.  
It can identify the pattern in the relationship between the two variables,  
Median Household Income and Opioiod Overdoses.  """

fig,ax = plt.subplots(nrows=1,ncols=2)

fig.suptitle('Virginia County Opioid rate Median Household Income & Opioid and Heroine/Fentanyl Overdose Relationship') # Graph Title
fig.set_size_inches(10,5)
ax[0].scatter(df_OpiodsVA['MedianHouseholdIncome'], df_OpiodsVA['FPOO-Rate'], alpha = .5, c = 'g')
ax[0].xaxis.set_label_text('Median Household Income')  # x-axis caption
ax[0].yaxis.set_label_text('Opioid Overdoses (per 100,000 county residents)')  # y-axis caption
ax[0].set_xlim(20000,150000)  # Set min and max for x axis
ax[0].set_ylim(-1,45)  # Set min and max for y axis 

ax[1].scatter(df_OpiodsVA['FPOO-Rate'], df_OpiodsVA['FFHO-Rate'], alpha = .5, c = 'b')
ax[1].xaxis.set_label_text('Opioid Overdoses (per 100,000 county residents)')  # x-axis caption
ax[1].yaxis.set_label_text('Heroine/Fentanyl Overdoses (per 100,000 county residents)')  # y-axis caption
ax[1].set_xlim(-1,45)  # Set min and max for x axis
ax[1].set_ylim(-1,45)  # Set min and max for y axis 
plt.show()

"""Questions:
o Do opioid overdoes tend to be associated with less affluent areas—that is, areas where
families have lower incomes?
    The scatter plot shows a strong relationship between low median income households and opioid overdoses.  As median household
    income increases the number of overdoses decreases. 
o Some people who start with opioid addictions are reported to transition to heroin use.
What is the relationship in Virginia counties between opioid overdoses and heroin
overdoses?  
    There does not seem to be a correlation between the number of opioid overdoses and number of heroine/fentanyl overdoses
    in the second graph."""
```

```python
df_cars = pd.read_csv('TOTALNSA.csv')

"""A line graph is used to plot continuous data and 
the connection of points is necessary for time-series data."""

fig,ax = plt.subplots()

fig.suptitle('Historical US Vehicle Sales Trend') #Graph Title
fig.set_size_inches(12,5)
dates = mdates.datestr2num(df_cars['DATE'])
ax.plot(dates,df_cars['TOTALNSA'], color='k')
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
ax.xaxis.set_label_text('Date (Monthly Data)')  # x-axis caption
ax.yaxis.set_label_text('Total US Vehicle Sales ($ per Month)')  # y-axis caption
ax.set_ylim(0,2000)  # Set min and max for y axis

plt.show()

"""Question:
o How have vehicle sales in the United States varied over time?
    Overtime, total vehicle sales in the United States have been in a range according to this data.  
    However, in recent years we can see that it has been steadily increasing.  There also appears to be
    seasonality in the data.  """
```
